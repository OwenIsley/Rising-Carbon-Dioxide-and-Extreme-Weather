```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
df = pd.read_excel('PercentChange.xlsx')
del df['Year']
df
```

```python
df *= 100
```

```python
df['Percent change carbon']= round(df['Percent change carbon'],1)
df['Percent change damage']= round(df['Percent change damage'],-1)
df['Percent change injuries']= round(df['Percent change injuries'],-1)
df['Percent change number of storms']= round(df['Percent change number of storms'],-1)
df.tail()
```

```python
df['Percent change carbon'] = 'C' + df['Percent change carbon'].astype(str)
df['Percent change damage'] = 'D' + df['Percent change damage'].astype(str)
df['Percent change injuries'] = 'I' + df['Percent change injuries'].astype(str)
df['Percent change number of storms'] = 'S' + df['Percent change number of storms'].astype(str)
df.head()
```

```python
records = []
for i in range(0, 52):
    records.append([str(df.values[i,j]) for j in range(0, 4)])
```

```python
association_rules = apriori(records, min_support=0.04, min_confidence=0.5, min_lift=.1, min_length=2)
association_results = list(association_rules)
```
